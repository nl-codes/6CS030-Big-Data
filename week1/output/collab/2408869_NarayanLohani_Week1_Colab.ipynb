## Exercise on Loops
**Question 1:** Display **Fibonacci series** up to 10 terms.



```python
# Fibonacci Series
upto = 10
first = 0
second = 1

# Printing first two numbers
print("Fibonacci series up to", upto, "terms:")
print(first, second, end=" ")
for i in range(upto - 2):
    next = first + second
    print(next, end=" ")
    # Updating values for next iteration
    first = second
    second = next
```

**Question 2:** Find the factorial of a given number. Ask the user to provide the number.

```python
# Factorial of user given number
user_input = int(input("Enter number to calculate factorial: "))
num = user_input
factorial = 1

# Caluclation factorial as : n * (n-1) * (n-2) * ... * 2
for i in range(2, num + 1):
    factorial *= i

print(f"Factorial({user_input}) = {factorial}")
```

**Question 3:** Write a program to print a right angle triangle using the for loop.

```python
# Right angle triangle pattern using '*'

# Prompt user for height of the triangle
height = int(input("Enter height of the right angle triangle : "))

for i in range(1, height + 1):
    for j in range(i):
        print("*", end=" ")
    print()  # New line afer each row
```

## Exercise on Functions
**Question 1:** Make a simple calculator that can add, subtract, multiply and divide two numbers using function.

```python
def user_choice():
    """
    Display calculator menu and get operation choice from user.

    Returns:
        int: The operation choice (1=Add, 2=Subtract, 3=Multiply), 0 if invalid.
    """

    print("-----Simple Calculator-----\n1. Add\n2. Subtract\n3. Multiply")
    choice = int(input("Enter choice: "))

    # Validate choice
    if choice not in [1, 2, 3]:
        print("Invalid operation choice")
        print("Exiting calculator...")
        return 0
    return choice


def get_user_input():
    """
    Prompt user for two operands.

    Returns:
        tuple: Two float numbers entered by the user.
    """
    first = float(input("Enter first operand: "))
    second = float(input("Enter second operand: "))
    return first, second


def calculate(first, second, operation):
    """
    Perform calculation based on the operation.

    Parameters:
        first (float): First operand.
        second (float): Second operand.
        operation (int): Operation code (1=Add, 2=Subtract, 3=Multiply).
    """
    match operation:  # python 3.10+ uses match instead of switch
        case 1:
            print(f"{first} + {second} = {first + second}")
        case 2:
            print(f"{first} - {second} = {first - second}")
        case 3:
            print(f"{first} * {second} = {first * second}")


def main():
    """
    Main function to run the calculator.
    """
    operation = user_choice()
    if operation == 0:
        return

    first, second = get_user_input()
    calculate(first, second, operation)


if __name__ == "__main__":
    main()
```

## Exercise on List and Dictionaries

**Question 1:** Remove the first duplicate element of a list

```python
def remove_first_duplicate(numbers):
    """
    Remove the first duplicate element from the list.

    Parameters:
        numbers (list[int]): List of numbers.
    Returns:
        list[int]: List after removing the first duplicate.
    """
    non_duplicates = set()
    for idx, ele in enumerate(numbers):
        if ele in non_duplicates:
            print("First Duplicate Element : ", ele)
            numbers.pop(idx)
            return numbers
        non_duplicates.add(ele)
    return numbers


given_list = [1, 2, 4, 1, 5, 2]
print("Original list : ", given_list)
removed_ele_list = remove_first_duplicate(given_list)
print("List after removing the first duplicate item : ", removed_ele_list)
```

**Question 2:** Write a Python program to create a new dictionary by extracting the mentioned keys from the below dictionary.

>sample_dict = { "name": "Anderson", "age": 35, "salary": 5000, "city": "London"}

>keys = ["name", "salary"]

>**Expected Output: {'name': 'Anderson', 'salary': 5000}**


```python
to_extract_keys = ["name", "salary"]
given_dict = {
    "name": "Kuchiki Byakuya",
    "age": "20",
    "salary": "1000000",
    "city": "Seretie",
}
print("Original dictionary:\n", given_dict)

new_dict = dict()
for dict_key, dict_value in given_dict.items():
    if dict_key in to_extract_keys:
        new_dict[dict_key] = dict_value

print("Dictionary after extracting keys:\n", new_dict)
```

**Question 3:** Write a Python program to convert two lists into a dictionary in a way that item from list_one is the key and item from list_two is the value.

>list_one = ['one', 'two', 'three']

>list_two = [1, 2, 3]


>**Expected Output: {'one': 1, 'two': 2, 'three': 3}**

```python
list_keys = ["one", "two", "three"]
list_values = [1, 2, 3]

print("List of keys:", list_keys)
print("List of values:", list_values)

dict_numbers = dict()

for idx, ele in enumerate(list_keys):
    dict_numbers[ele] = list_values[idx]

print("Dictionary after combining lists:", dict_numbers)
```
